```python
bangla_stopwords = set([
    "আমি","আমরা","তুমি","আপনি","সে","তিনি","তাহারা","তারা","যে","যা","যাকে","যাদের",
    "এই","ওই","কোনো","সব","কিছু","কেউ","যখন","যদি","কেন","কি","কোথায়",
    "করব","করবে","করেন","করছি","করছে","করে","হবে","হয়","হই","ছিল","থেকে","থাকবে","থাকেন","থাকে",
    "আছে","আছ","আছি","নয়","না","হয়নি","হওয়া","যায়"
])
```

```python
def clean_bangla_text(text):
    text = str(text)
    text = re.sub(r"[^\u0980-\u09FF\s০-৯]", "", text)
    tokens = text.split()
    tokens = [w for w in tokens if w not in bangla_stopwords]
    return " ".join(tokens)
```

```python
text_data = "আমি বাংলাদেশে থাকি এবং আমি আমার দেশকে ভালোবাসি। আমার দেশ সুন্দর।"
```

```python
cleaned_text = clean_bangla_text(text_data)
print("Cleaned:", cleaned_text)
```

```python
font_path = "C:/Users/Jason/Downloads/kalpurush.ttf"
```

```python
wordcloud = WordCloud(
    font_path=font_path,  # Full path to TTF
    background_color="white",
    width=800,
    height=600
).generate(cleaned_text)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
